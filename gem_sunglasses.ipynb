```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
```

```python
def scrape_gem_opticians():
    output_file = "gem_sunglasses.csv"

    with open(output_file, mode="w", encoding="utf-8") as fd:
        fd.write("Brand,Code and Color,Price\n")

        options = Options()
        options.add_argument('--headless')
        driver = webdriver.Chrome(options=options)

        for n in range(56):  # Number of pages to scrape
            base_url = f"https://gemopticians.com/collections/mens-sunglasses?page={n}"
            driver.get(base_url)
            time.sleep(5)

            soup = BeautifulSoup(driver.page_source, "html.parser")

            brands = soup.find_all("p", class_="ProductItem__Vendor Heading")
            titles = soup.find_all("h2", class_="ProductItem__Title Heading")
            prices = soup.find_all("span", class_="money")

            for i in range(min(len(brands), len(titles), len(prices))):
                brand = brands[i].text.strip()
                title = titles[i].find("a").text.strip()
                price = prices[i].text.strip()

                fd.write(f"\"{brand}\",\"{title}\",\"{price}\"\n")

            print(f"✅ Page {n + 1} done")

        driver.quit()

    print("\n✅ Scraping completed successfully!")


def preview_csv():
    df = pd.read_csv("gem_sunglasses.csv")
    print(df.head())
```

```python
if __name__ == "__main__":
    scrape_gem_opticians()
    preview_csv()
```
